```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
data=pd.read_csv(r"C:\Users\adity\Documents\6 sem tiet\ML\sonar.csv")
data=np.array(data)
data_samples=data[:,:-1]
data_target=data[:,-1]
rocks=np.zeros(96)
mines=np.ones(111)
target=np.concatenate((rocks,mines),axis=0)
df=pd.DataFrame(data_samples)
df['target']=target

from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,:-1],df['target'],test_size=0.2,random_state=0)
accu=[]
for n in range(1,11):
    knn=KNeighborsClassifier(n_neighbors=n)
    knn.fit(X_train,y_train)
    pred=knn.predict(X_test)
    accuracy=accuracy_score(y_test,pred)
    accu.append(accuracy)
plt.plot(range(1,11),accu)
plt.xlabel("test size")
plt.ylabel("accuracy")
plt.title("KNN")
```

```python
from sklearn.tree import DecisionTreeClassifier
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,:-1],df['target'],test_size=0.2,random_state=0)
accu=[]

for n in range(1,11):
    tree=DecisionTreeClassifier(max_depth=n)
    tree.fit(X_train,y_train)
    pred=tree.predict(X_test)
    accuracy=accuracy_score(y_test,pred)
    accu.append(accuracy)
plt.plot(range(1,11),accu)
plt.xlabel("test size")
plt.ylabel("accuracy")
plt.title("Decision Tree")
```
